### **ACELERANDO CON RE.COMPILE**

Existen otros metodos, otras busquedas y formas de realizar estas busquedas utilizando REGEX pero de una forma mas productiva, mas rapida.

```python
# 1°) Metodo SEARCH() de la biblioteca RE de REGEX, que realiza busquedas de acuerdo un REGEX --->

# Ejemplo: Buscamos el 70 que sera convertido en una REGEX y va a ser buscado dentro de la
# cadena de texto:
re.search(r'70', '12345678901234567890123456789012345670')

# Nos devuelve el objeto Match=re.Match object, posision donde se encuantra=span=(36, 38), y valor que fue buscado=match='70'
```

```python
# 2°) Ahora usamos el Metodo COMPILE(), nos ayuda a buscar REGEX, lo que hacemos es Compilar antes y esta compilacion la
# estamos guardando en una Variable y encima de esta variable que contiene Compile aplicamos SEARCH() --->

regex = re.compile(r'70') # Variable regex que contiene metodo Compile
regex.search('12345678901234567890123456789012345670')

# Tenemos el mismo resultado, tenemos 2 formas de buscar con REGEX dentro de nuestra cadena de texto.
```

```python
# 3°) Vamos a ver cual es la diferencia de SEARCH() y COMPILE() con la biblioteca TIMEIT que nos muestra el tiempo
# de ejecucion de los metodos ---->

# Importamos timeit:
from timeit import timeit

setup = '''import re''' # Creamos varible e importamos la biblioteca, las ''' para avisar que vamos a usar codigo.
timeit('''re.search(r'70', '12345678901234567890123456789012345670')''', setup)

# Nos devuelve el tiempo que demoro en ejecutar el codigo con el metodo SEARCH que fue de 0.68 segundos.
```

```python
# Ahora hacemos lo mismo con el metodo COMPILE:

setup = '''import re
regex = re.compile(r'70')''' # Pasamos tambien la variable REGEX que contiene el metodo Compile
timeit('''regex.search('12345678901234567890123456789012345670')''', setup)

# El tiempo fue de 0.19 segundos, mas rapido que el codigo anterior, COMPILE es mucho mas rapido!
```

### **REGEX: ELIMINANDO TAG HTML**

* **En el Metodo SEARCH()** ---> cada vez que utilizamoes este metodo, la busqueda r'70' es transformada en una REGEX y tambien se esta compilando en ese mismo momento el STRING a una REGEX, esto genera un tiempo adicional hasta obtener el resultado, cada vez que genero la llamada 1.000.000 de veces r'70' se genera 1.000.000 de veces la compilacion de STRING a REGEX.

* **En el Metodo COMPILE()** ---> En este metodo primero se Compila el STRING r'70' para convertirlo en REGEX una unica vez antes de realizar el 1.000.000 de llamada, por lo tanto, cuando luego realizo el metodo SERACH() directamente de la REGEX ya compilada, cuando realize el 1.000.000 de llamadas ya NO se esta transformando la REGEX ya que se hizo anteriormente y esto lo vuelve mas rapido.

```python
# 1°) Ahora vamos a usar el METODO COMPILE() dentro de una Funcion Eliminar --->

def eliminar(textos, regex): # Recibe TEXTOS que es la Pregunta y REGEX ya compilada
# Diferenciamos de un texto a mas de un texto, en este caso el if= si el tipo de taxto es STRING, es texto
# de una LINEA!:
  if type(textos) == str:
# Devuelve la REGEX y utilizamos el metodo SUB()=sustituye con vacio ' ' cada vez que encuentres las Etiquetas
# y recibe TEXTO tambien:
    return regex.sub(' ', textos)
# Cuando se mas de un texto, una LISTA de textos, cuando recibimos una tabla completa:
  else:
# Devolvemos una LISTA[] y sustituimos con texto= los texto de cada una de las filas que reciba
# en mi lista de textos:
    return [regex.sub(' ', texto) for texto in textos] # texto= cada linea
```

```python
# 2°) Ahora generamos nuestra REGEX ---->

regex_html = re.compile(r'<.*?>')
# Generamos variable Pregunta sin Etiqueta:
pregunta_sin_tag = eliminar(pregunta_portugues, regex_html) # Usamos la Funcion Eliminar y le pasamos los datos de Pregunta_Portuges y el REGEX con COMPILE.
print(pregunta_sin_tag)

# Nos devuelve la Pregunta sin los Tag(Etiquetas)!, verifiacamos con Pregunta_español y Pregunta_ingles.
```

```python
# Verificamos en pregunta_español:

pregunta_sin_tag = eliminar(pregunta_español, regex_html)
print(pregunta_sin_tag)
```

```python
# Verificamos en pregunta_ingles:

pregunta_sin_tag = eliminar(pregunta_ingles, regex_html)
print(pregunta_sin_tag)
```

### **ELIMINANDO PEDAZOS DE CODIGO**

Ahora nos vamos a encargar de borrar los CODIGOS y para eso vamos a crear una Funcion.

```python
# 1°) Creamos la Funcion para eliminar los CODIGOS y vamos a reeutilizar el codigo de la Funcion Eliminar y cambiamos el nombre
# de la Funcion ---->

# Vamos a sustituir CODIGO y vamos a recibir una Lista de texto y una Regex diferente a la cual veniamos
# utilizado para las Etiquetas:
def substituir_codigo(textos, regex):
  if type(textos) == str:
    return regex.sub('CODE', textos) # Vamos a sustituir por la palabra CODE en donde este el codigo
  else:
    return [regex.sub('CODE', texto) for texto in textos]
```

```python
# 2°) Examinamos alguna Pregunta en español para ver como vamos a crear nuestra REGEX para sustituir los codigos por
# code --->

print(pregunta_español)

# Podemos observar que el codigo viene acompañado al principio por la palabra <code> inicialmente y </code> en el final del codigo.
```

```python
# 3°) Creamos REGEX, lo compilamos con el metodo COMPILE --->

regex_codigo = re.compile(r'<code>(.|(\n))*?</code>')
# (.|(\n))= luego de code, busca cualquier caracter(.) o(|) salto de linea(\n) y 1 palabra o infinitas palabras o letras(*)
#  desde el primer code hasta el ultimo code de la linea(?)
```

```python
# 4°) Probamos la REGEX generada en alguna Pregunta en los diferente idiomas para ver si funciono, para esto creamos Variable --->

pregunta_sin_code = substituir_codigo(pregunta_español, regex_codigo)
print(pregunta_sin_code)

# Nos devuelve la pregunta con los codigos sustituidos con CODE.
```

```python
# Hacemos lo mismo con portugues e ingles --->

pregunta_sin_code = substituir_codigo(pregunta_portugues, regex_codigo)
print(pregunta_sin_code)

# Funciono
```

```python
pregunta_sin_code = substituir_codigo(pregunta_ingles, regex_codigo)
print(pregunta_sin_code)
```

### **ELIMINACION DE DIGITOS Y PUNTUACIONES**

Como notamos que en algunos textos que tenemos signos de pregunta(?) y el objetivo es dejar solo caracteres Alfanumericos en el texto y para que el modelo pueda trabajar solamente con texto debemos eliminar ?,!,., numeros.

```python
# 1°) Creamos REGEX para tratar las puntuaciones --->

regex_puntuacion = re.compile(r'[^\w\s]') # ^ = Excluye lo que esta despues de ^ para que NO sea eliminado, en este caso el Texto y espacios.
print(eliminar(datos_español.sin_code_tag[0], regex_puntuacion))#Pasamos la Funcion Eliminar a la linea ya limpia y el REGEX

# Nos elimina los simbolos y deja los caracteres alfanumericos y espacios.
```

```python
# 2°) Podemos ver que tenemos letras en MAYUSCULA y MINUSCULAS que son la misma palabra y el modelo puede tener un Error de interpretacion
# debemos normalizar a Minuscula, para esto vamos a utilizar la funcion que ya tenemos preparada mas arriba --->

def minuscula(textos): # NO necesitamos la REGEX porque lo unico que va hacer es recibir el texto y devolver texto en minuscula.
  if type(textos) == str:
    return textos.lower() # Cada vez que reciba un texto que nos retorne el mismo texto en minuscula.
  else: # Si recibo varios textos que los devuelva en minuscula
    return [texto.lower() for texto in textos]
```

```python
# 3°) Vemos si la Funcion funciona bien y nos devuelve minuscula ---->

print(minuscula(datos_español.sin_code_tag[0]))

# Nos devuelve las 2 "las" en minuscula.
```

### **ESPACIOS DUPLICADOS**

Tenemos que tratar los ESPACIOS y SALTOS DE LINEA entre las palabras.

```python
# 1°) Generamos la REGEX para el salto de linea y para el espacio --->

regex_salto_linea = re.compile(r'(\n)') # Salto de linea entre () porque sino regex lo toma como texto.
regex_espacio = re.compile(r' +') # 1 espacio o mas (+) de una espacio
```

```python
# 2°) Ahora hacemos la Funcion, reutilizamos la Funcion Substituir pero el lugar de sustituir con CODE lo hacemos con el
# espacio ---->

def substituir_por_espacio(textos, regex):
  if type(textos) == str:
    return regex.sub(' ', textos) # Vamos a sustituir por espacio
  else:
    return [regex.sub(' ', texto) for texto in textos]
```

```python
# 3°) Ahora vemos si obtuvo el resultado esperado ---->

resultado = (substituir_por_espacio("Alura 1234  \n\n    Latam 123", regex_salto_linea))
print(substituir_por_espacio(resultado, regex_espacio)) # Utilizamos el resulatdo de la variable con el salto de linea y le agregamos regex de espacio.

# Nos devuelve el texto sin salto de linea(\n) ni espacios.
```

### **PIPELINE PARA ENTRENAMIENTO DE ML**

Ya separamos los datos en Entrenamiento y Prueba, 400 PREGUNTAS separadas en un objeto SERIES de Entrenamiento y 100 PREGUNTAS separadas en un objeto SERIES de Prueba. Lo que tenemos que hacer es juntar estas 400 Preguntas en un gran texto, es decir CONCATENAR en una unica Variable y para esto vamos a realizar una Funcion.

```python
# 1°) Realizamos una Funcion que va a CONCATENAR las 400 Preguntas de Entrenamiento ---->

todas_palabras_port = ' '.join(port_train) # Creamos la Varible, colocamos espacio ' ' antes del JOIN= que Concatena las 400 Preguntas.
print(todas_palabras_port)

# Nos devuelve todas las PREGUNTAS(400) CONCATENADAS  del Entrenamiento en Portugues.
```

```python
# 2°) Ahora tenemos que separar el TEXTO CONCATENADO en caracteres de a 2 con BIGRAMS, tambien debemos crear una Variable que separe tambien
# palabra por palabra el texto y por ESPACIO, para esto usaremos el Metodo de Procesamiento de Lenguaje Natural NLTK Whitespace ----->

from nltk.tokenize import WhitespaceTokenizer # Separar las palabras por ESPACIO en blanco.

todas_palabras_port = WhitespaceTokenizer().tokenize(todas_palabras_port) # Separamos las palabras del texto
print(todas_palabras_port)

# Nos devuelve las palabras separadas.
```

```python
# 3°) Ahora entramos a cada una de las palabras y separamos en BIGRAMS y colocamos el CARACTER FALSO(Fake Chars) al inicio
# y al final de cada palabra, para esto vamos a usar otra Biblioteca parecida a la que usamos anteriormente PADDED_EVERYGRAM_PIPELINE
# de NLTK que genera Bigrmas y ademas genera vocabulario con Frecuencia que nos perimite saber cuantas veces aparace la combinacion
# por ejemplo de la letra a con b, nos devuelve Estadistica ---->

from nltk.lm.preprocessing import padded_everygram_pipeline

# Variable BIGRAMS, Variable Vocabulario Frecuencia = metodo con 2 caracteres y palabras separadas:
port_train_bigram, vocab_port = padded_everygram_pipeline(2, todas_palabras_port)

# Para poder ver si esta generado en formato BIGRAMS:
next(next(port_train_bigram))
```

### **CREANDO EL MODELO MLE**

Ya tenemos los datos de BIGRAMS y vamos a construir nuestro Modelo , si vamos a la documentacion de NLTK podemos ver que tenemos informacion sobre diversos modelos que pueden ser aplicados y construidos con la biblioteca NLTK, el que usaremos es el MLE, Mayor Estimados de Probabilidad, este modelo nos va a ayudar a estimar la Mayor Probabilidad en la combinacion de algunas palabras para que escoja el IDIOMA mas Probable en el cual se ha escrito el texto.

```python
# 1°) Importamos la biblioteca MLE, construimos el modelo y entrenamos el modelo ---->

from nltk.lm import MLE

modelo_port = MLE(2) # Creamos variable, cantidad de Bigrams = 2
# Entrenamos el modelo:
modelo_port.fit(port_train_bigram, vocab_port) # datos Bigrams y vocabulario Frecuencia.
```

```python
# 2°) Para ver si el Entrenamiento del modelo con MLE esta funcionando podemos usar .GENERATE que genera palabras de la
# cantidad de digitos que querramos(NUM_WORDS = 6), como nosotros separamos en caracteres va a generar palabras con
# 6 caracteres de toda la base de Preguntas del texto unido de Portugues ---->

modelo_port.generate(num_words = 6) # Selecciona la Secuencia con mayor Probabilidad segun los datos que hemos pasado.

# Existen muchas palabras que Problablemente empiezen con la letra E y luego le siga la S ... etc. (en cada ejecuacion cambia.)
```

```python
# 3°) Vemos la Frecuencia de letras con la Variable VOCAB_PORT y para esto usamos un biblioteca de NLTK NGRAMCOUNTER, importamos --->

from nltk.lm import NgramCounter

modelo_port.counts[['e']].items() # Frecuencia luego de la letra 'e' que nos devolvio arriba.

# Nos devuelve luego de caracteres falso('</s>', 7489), la Probabilidad de que luego de la letra 'e' le siga la letra('s', 2798).
```

### **PERPLEJIDAD PARA PALABRAS**

Ya tenemos nuestro Modelo construido , ya hemos alimentado nuestro Modelo con los datos en PORTUGUES separados en formato BIGRAMAS, entrenado y hemos pasado algunas palabras generando 6 caracteres que tienen una Secuencia de Probabilidad segun los  testo que hemos alimentado nuestro modelo MLE.

```python
# 1°)Ahora vamos a colocar una Frase para ver si es reconocido el idioma Portugues con el modelo y en otros idiomas
# para ver si el Modelo lo rechaza ---->

# Creamos frase en Portugues:
texto = "bom dia"
# Separamos las palabras en un un lista:
palabras = WhitespaceTokenizer().tokenize(texto) # Separamos por espacion en blanco con Tokenize.
# Creamos los Caracteres Falsos:
palabras_fakechar = [list(pad_both_ends(palabra, n=2)) for palabra in palabras] #palabra=pasamos palabra por palabra y luego recorremos Variable Palbras.
# Creamos BIGRAMS:
palabras_bigrams = [list(bigrams(palabra)) for palabra in palabras_fakechar]

print(palabras_bigrams)

# Nos devuelve las 2 plabras bom dia separadas con caracteres falsos al inicio y final y en bigrams.
```

### **FUNCION PARA ENTRENAR ML'S**

Nosotros queremos detectar TEXTO como lo hace el detector de GOOGLE, para traducir texto tenemos que crear una Funcion que este preparada para recibir TEXTO como parametro y devolvernos el Modelo Entrenado con estos Textos, Funcion Programada para leer grandes Textos.

```python
# 1°) Creamos la Funcion, que reciba una lista en forma de columna de TEXTO --->

def entrenar_modelo_mle(lista_textos): # Recibe una lista de texto.
  todas_preguntas = ' '.join(lista_textos) # Concatena lista Texto unica Variable.
  todas_palabras = WhitespaceTokenizer().tokenize(todas_preguntas) # Separamos por Espacio la Variable del texto Concatenado.
  bigrams, vocabulario = padded_everygram_pipeline(2, todas_palabras) # Variables bigrams, vocabulario, padded_everygram_pipeline=devuelve FakeChars y Bigrams.
  modelo = MLE(2) # Creamos el MODELO
  modelo.fit(bigrams, vocabulario) # Entrenamos el modelo.

  return modelo
```

### **PERPLEJIDAD PARA FRASES**

Ahora lo que haremos es calcular la PREPLEJIDAD para una Frase completa , NO para cada palabra separada como GOOD/MORNING, calcular el TEXTO completo y su precision par aver si esa frase esta adecueda a un modelo o a otro de otro idioma.






```python
# 1°) Creamos la Funcion para Calcular la PERPLEJIDAD --->

# Parametro Modelo y Texto para preguntarle a la maquina  si el texto
# que estoy pasando esta de acuerdo al modelo de inglea o español:
def calcular_perplejidad(modelo, texto):
  perplejidad = 0 # Contador que inicia en 0.
  palabras = WhitespaceTokenizer().tokenize(texto)
  palabras_fakechar = [list(pad_both_ends(palabra, n=2)) for palabra in palabras]
  palabras_bigrams = [list(bigrams(palabra)) for palabra in palabras_fakechar]

  for palabra in palabras_bigrams: # Recorre cada palabra en las palabras Bigrams.
    perplejidad += modelo.perplexity(palabra) # Aumentamos el contador para que sume la perplejidad de cada palabra recorrida

  return perplejidad
```

### **PERPLEJIDAD INFINITA**

Nos encontramos con un obstáculo, pero eso es lo bueno. Eso es lo bueno de Data Science. Tener obstáculos generan desafíos. ***¿Cuál es el obstáculo?*** Nosotros recibimos una perplejidad, cuando estábamos calculando la perplejidad, ***en lugar de recibir un número como era esperado, nosotros recibimos un valor textual, escrito inf.***

***Esto significa perplejidad infinita y cuando tenemos perplejidad infinita no podemos calcular pues la perplejidad y como no podemos calcular la perplejidad, no sabemos si el modelo, si la palabra o el texto que estamos colocando corresponde a un modelo o no, necesitamos un número para poder evaluar.***

¿Ahora qué significa perplejidad infinita? Alejandro, ¿Por qué no se puede tener un resultado normal como los otros? Buena pregunta, vamos a la solución. Primero tuvimos que el resultado de perplejidad infinita, cuando estábamos utilizando el modelo en español.

***El modelo español fue entrenado con palabras en español, lógicamente entonces, el modelo solo espera palabras en español, pero nosotros no le mandamos un texto en español. Nosotros le mandamos esto de aquí, este texto y este texto está en portugués.***

Entonces, lo que quiere decir perplejidad infinita es que al momento de enviarle a nuestro modelo, que está escrito en español, este texto que está escrito en portugués, alguna palabra, algún carácter, algún bigram de aquí, alguna tupla, algún par de caracteres aquí no existía en el modelo en español, por ejemplo, podría imaginar que estos dos caracteres nunca, pero nunca serán encontrados en las palabras en español que yo le pasé al modelo en español, al momento de entrenarlo.

Porque estos son caracteres específicos del idioma en portugués. Entonces, cuando el modelo recibe palabras que no existían en su entrenamiento, regresa como cálculo de la perplejidad, regresa un valor infinito. ¿Y por qué un valor infinito? Porque la probabilidad de que estos dos caracteres o caracteres existan en español es cero probabilidad cero de que estos caracteres existen. ¿Y cómo se calcula la perplejidad?

Pues la perplejidad es lo inverso de la probabilidad, en la teoría nos dice que la perplejidad es lo inverso de la probabilidad. Entonces, si es lo inverso de la probabilidad y 1 es dividido por la probabilidad, en este caso, probabilidad de que existan estos dos caracteres es 0, entonces el cálculo va a hacer 1 sobre 0, y todo lo que sea dividido entre 0 nos va a dar como resultado infinito, va a tender al infinito.
Es tan pequeño el resultado que es probable que el resultado se marque como infinito. Entonces, cuando la probabilidad es 0, el cálculo de la perplejidad va a ser infinito y como estamos sumando aquí en nuestra función, como estamos sumando perplejidades de cada palabra en el momento que alguna de ellas sea infinita, entonces todo el resultado será infinito y va a regresar reinito en nuestra función.

***Ese es el problema, el obstáculo que encontramos al ejecutar nuestro modelo MLE con palabras que no existen.*** ¿Ahora cómo podemos salir nuestro obstáculo? Ese es el desafío, ¿cómo podemos superar esto? Existen otros modelos dentro de la biblioteca NLTK que desconsideran o que trabajan bien con palabras que no existían al momento del entrenamiento, por lo menos no va a devolver infinito en el cálculo de la perplejidad.

Existen otros métodos de tratar esto y para eso tenemos que utilizar uno de ellos, por ejemplo, el modelo de Laplace. ***El modelo de Laplace trabaja bien cuando la probabilidad de que no exista algún carácter es 0, entonces este es el modelo que vamos a continuar trabajando a partir de ahora, porque el modelo MLE ya vimos que no va a funcionar bien en estos casos.***

Entonces vamos a cambiar ahora para un mejor modelo también de la biblioteca NLTK llamado modelo de Laplace.

### **MODELO DE LAPLACE**

Tenemos PERPLEJIDAD INFINITA debido a que NO se encuentrasn palabras en PORTUGUES en un modelo que fue previamente entrenado en ESPAÑOL y esto genera Perplejidad infinita 0(INF), NO podemos obtener un numero y nuestro modelo NO podra saber si las palabras o el texto pertenece a Portugues, Ingles o Español. Y para esto vamos a trabajar con el MODELO LAPLACE, creamos otro modelo que cuando este modelo este ejecutandose y encuentre Perplejidad Infinita 0 va a intentar suavizar el resultado para entregarnos un valor bajo para que NO afecte el resultado Final.

```python
# 1°) Importamos el modelo, copiamos la Funcion la cual creamos anteriormente para Entrenar el Modelo MLE y en lugar de Entrenar
# este modelo vamos a entrenar el modelo LAPLACE ---->

from nltk.lm import Laplace

def entrenar_modelo_Laplace(lista_textos): # Cambiamos a modelo LAPLACE.
  todas_preguntas = ' '.join(lista_textos) #
  todas_palabras = WhitespaceTokenizer().tokenize(todas_preguntas)
  bigrams, vocabulario = padded_everygram_pipeline(2, todas_palabras)
  modelo = Laplace(2) # Creamos el MODELO LAPLACE
  modelo.fit(bigrams, vocabulario) # Entrenamos el modelo.

  return modelo
```

```python
# 2°) Ahora ENTRENAMOS  los 3 modelo con el MODELO LAPLACE --->

modelo_port_Laplace = (entrenar_modelo_Laplace(port_train))
modelo_esp_Laplace = (entrenar_modelo_Laplace(esp_train))
modelo_ing_Laplace = (entrenar_modelo_Laplace(ing_train))
```

```python
# 3°) Ahora vamos a ver si nos devuelve PERPLEJIDAD INFINITA o NO --->

# Usamos nuevamente la Funcion:
def calcular_perplejidad(modelo, texto):
  perplejidad = 0 # Contador que inicia en 0.
  palabras = WhitespaceTokenizer().tokenize(texto)
  palabras_fakechar = [list(pad_both_ends(palabra, n=2)) for palabra in palabras]
  palabras_bigrams = [list(bigrams(palabra)) for palabra in palabras_fakechar]

  for palabra in palabras_bigrams: # Recorre cada palabra en las palabras Bigrams.
    perplejidad += modelo.perplexity(palabra) # Aumentamos el contador para que sume la perplejidad de cada palabra recorrida

  return perplejidad

# Modelo Portugues, palabras entrenamiento Portugues:
print(calcular_perplejidad(modelo_port_Laplace, port_test.iloc[0]))
```

```python
# Ahora volvemos a calcular donde nos dio INFINITO --->

# Modelo Español, palabras entrenamiento Portugues:
print(calcular_perplejidad(modelo_esp_Laplace, port_test.iloc[0]))

# Ya NO tenemos INF, aunque la palabra NO exista dentro de los datos de Entrenamiento en modelo Español, NO vamos a tener Probabilidad 0,
# porque el modelo LAPLACE esta evitando que tengamos estos casos de INF.
```

```python
# Ahora lo calculamos con modelo Ingles --->

# Modelo Ingles, palabras entrenamiento Portugues:
print(calcular_perplejidad(modelo_ing_Laplace, port_test.iloc[0]))

# Nos da un resultado en numero y NO en INF.
```

### **ATRIBUIR IDIOMAS**

Vamos a generar una FUNCION que nos permita obtener como resultado el nombre del Idioma al cual corresponde el Texto y NO numeros.

```python
# 1°) Vamos a crear la Funcion ASIGNAR IDIOMA y le pasamos como parametro nuestra LISTA_TEXTOS ---->

def asignar_idioma(lista_textos):
# Generamos una lista vacia para poder colorcar los nombres de los Idiomas que va identificando:
  idiomas = []
# Iteramos con un FOR los texto de los textos que tenemos en lista_textos:
  for texto in lista_textos:
# Creamos la Variable Portugues para calcular la PERPLEJIDAD del modelo en Portugues de LAPLACE para
# cada pregunta que esta entrendo, queremos saber si la pregunta que esta entrando es escrita en Portugues:
    portugues = calcular_perplejidad(modelo_port_Laplace, texto)
# Hacemos lo mismo con los otros 2 idiomas, para obetner tambien los numeros:
    ingles = calcular_perplejidad(modelo_ing_Laplace, texto)
    español = calcular_perplejidad(modelo_esp_Laplace, texto)

# Una vez que tenemos los numeros hacemos un IF:
    if ingles >= portugues <= español: # Si el valor Portugues es el menor de los 3 idiomas quiere decir que el idioma esta en Portugues.
# Y lo adicionamos a la lista vacia:
      idiomas.append("portugues")
    elif portugues > ingles < español: # Si el ingles es el menor de lo3 3 idiomas.
# Lo adicionamos a la lista vacia:
      idiomas.append("ingles")
# Si no cumple las 2 condiciones de arriba:
    else:
      idiomas.append("español")

  return idiomas  # Retornamos la lista idiomas.
```
